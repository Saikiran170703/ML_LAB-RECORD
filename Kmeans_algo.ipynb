```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv("Iris.csv")

iris.head()
```

```python
# x = iris.iloc[:,1:5]

x = iris.drop(columns=["Id","Species"])

x.head()
```

```python
# K Means

# Deciding Value of K
# The most crucial aspect of K-Means clustering is deciding the value of K
# We do this by performing Elbow Analysis 




wcss = []
k_values = list(range(1, 10))

for k in k_values:
    kmeans = KMeans(n_clusters=k, max_iter=100)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    # print("For cluster = {}, SSE/WCSS is {}".format(k, kmeans.inertia_))

plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

```python
# Running K-Means Model

# We now run K-Means clustering for obtaining a 3 cluster solution.


kmeans = KMeans(n_clusters=3, max_iter=100)
kmeans.fit(x)
iris["cluster"] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Sample points (3 clusters)
x_axis = np.array([
    1.8, 2.2, 2.0, 1.7, 2.3,       # Cluster 1
    7.8, 8.2, 8.0, 7.9, 8.3,       # Cluster 2
    4.8, 5.1, 5.0, 4.9, 5.2        # Cluster 3
])

y_axis = np.array([
    2.2, 1.9, 2.1, 2.3, 1.8,       # Cluster 1
    3.2, 3.1, 2.8, 3.0, 3.3,       # Cluster 2
    8.2, 8.3, 7.9, 8.1, 7.8        # Cluster 3
])

# Create a DataFrame
data = pd.DataFrame({'x': x_axis, 'y': y_axis})

# Plot the data
plt.figure(figsize=(6,6))
plt.plot()
plt.title('Dataset with 3 Natural Clusters')
plt.scatter(x_axis, y_axis)
plt.grid()
plt.show()
```

```python
print('For ONE Cluster, all data points should be assigned to 1 Cluster\n')

kmeans = KMeans(n_clusters = 1, max_iter = 100)
kmeans.fit(data)
data['clusters'] = kmeans.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```

```python
# Plot the data
plt.figure(figsize=(6,6))
plt.plot()
plt.title('Model when 1 cluster is given as input')
plt.scatter(x_axis, y_axis,c=data["clusters"])
plt.grid()
plt.show()
```

```python
print('For Two Cluster, all data points should be assigned to 2 Cluster\n')

kmeans = KMeans(n_clusters = 2, max_iter = 100)
kmeans.fit(data)
data['clusters'] = kmeans.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```

```python
# Plot the data
plt.figure(figsize=(6,6))
plt.plot()
plt.title('Model when 2 cluster is given as input')
plt.scatter(x_axis, y_axis,c=data["clusters"])
plt.grid()
plt.show()
```

```python
print('For Three Cluster, all data points should be assigned to 3 Cluster\n')

kmeans = KMeans(n_clusters = 3, max_iter = 100)
kmeans.fit(data)
data['clusters'] = kmeans.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```

```python
# Plot the data
plt.figure(figsize=(6,6))
plt.plot()
plt.title('Model when 3 cluster is given as input')
plt.scatter(x_axis, y_axis,c=data["clusters"])
plt.grid()
plt.show()

```

```python
# Deciding Value of K
# The most crucial aspect of K-Means clustering is deciding the value of K
# We do this by performing Elbow Analysis 




wcss = []
k_values = list(range(1, 10))

for k in k_values:
    kmeans = KMeans(n_clusters=k, max_iter=100)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    # print("For cluster = {}, SSE/WCSS is {}".format(k, kmeans.inertia_))

plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```
