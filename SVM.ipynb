```python
# Prepare X and Y
X = iris_df.drop("Species", axis=1)
y = iris_df["Species"]
```

```python
#Split train and test data-70% train random state 1
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#Instantiate an object of class CSVSC using cost c=1
from sklearn.svm import SVC
model = SVC(C=1)
```

```python
#Fit the model
model.fit(X_train, y_train)
```

```python
# Predict using model
y_pred = model.predict(X_test)
```

```python
#Evaluate the model using the confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```

```python
# Print other metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1 Score:", f1_score(y_test, y_pred, average='macro'))
```

```python
# Print classification report and accuracy
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# K-fold cross validation
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)
print("K-Fold CV Scores:", scores)
print("Average CV Score:", scores.mean())
```

```python
#Use grid search CV to find optimal value of c-sel decision and lab questions
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000]}
grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(X, y)

print("Best parameters found:", grid.best_params_)
print("Best cross-validation score:", grid.best_score_)
print("Best estimator:", grid.best_estimator_)
```
